```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
#analysing profit across different regions
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Profit', data=df)
plt.title("Profit By Region")
plt.show()
```

```python
#Analysing profit across different categories
plt.figure(figsize=(10,6))
sns.barplot(x='Category',y='Profit',data=df)
plt.title('Profit by Category')
plt.show()
```

```python
#analysing proft across different sub-categories
plt.figure(figsize=(17,10))
sns.barplot(x='Sub-Category',y='Profit',data=df)
plt.title('Profit by sub-cateogry')
plt.show()
```

## Analysing Sales

```python
#analysing sales across different regions
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Sales',data=df)
plt.title('sales by region')
plt.show()
```

```python
#analysing sales by category
plt.figure(figsize=(10,6))
sns.barplot(x='Category',y='Sales',data=df)
plt.title('Sales by categories')
plt.show()
```
